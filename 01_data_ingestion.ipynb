## The Goal
Pull high-quality data for a diversified universe that reflects public equity focus (Large-cap, Financials, IT, and communication Services).

1. Define the Universe: We'll use the top 20 representative stocks from the S&P 500 across 5 key sectors (Tech, Finance, Healthcare, Energy, Industrials).

2. Download Data: Use yfinance to get 5 years of daily data.

3. Simulate the "TPM" Database: Structure the data in a "Long Format" (Tidy Data), which is what industrial SQL databases use.

```python
import yfinance as yf
import pandas as pd

# 1. Define diversified universe
tickers = [
    'AAPL', 'MSFT', 'GOOGL',  # Tech/Comm
    'JPM', 'GS', 'TD.TO',     # Financials (OMERS likes Canadian banks)
    'JNJ', 'PFE',             # Healthcare
    'XOM', 'CVX',             # Energy
    'CNR.TO', 'CP.TO',        # Industrials (Canadian Rails)
    'SHOP.TO', 'RY.TO'        # Growth/Quality
]
```

```python
#2. Download 5 years of data
data = yf.download(tickers, start="2021-01-01", end="2026-01-01", auto_adjust = False)['Adj Close']

#3. Transform to "Long Format" for the pipeline
df_long = data.stack().reset_index()
df_long.columns = ['Date', 'Ticker', 'Price']

print(f"Ingested {len(df_long)} rows of market data.")
df_long.head()
```
